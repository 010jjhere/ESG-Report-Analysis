```python
from datetime import datetime
```

```python
import yfinance as yf
import yahooquery as yq
```

```python
get_yq = yq.Ticker("aapl")
modules = 'assetProfile esgScores summaryDetail quoteType'
req_yq = get_yq.get_modules(modules)["aapl"]
```

```python
req_yq
```

```python
print("Symbol: AAPL")
```

```python
print("Peergroup:", req_yq['esgScores']['peerGroup'])
print("Environment Score:", req_yq['esgScores']["environmentScore"])
print("Social Score:", req_yq['esgScores']["environmentScore"])
print("Governance Score:", req_yq['esgScores']["governanceScore"])
print("Total Esg:", req_yq['esgScores']["totalEsg"])
print("Highest Controversy:", req_yq['esgScores']["highestControversy"])
print("Percentile:", req_yq['esgScores']["percentile"])
```

```python
print("Market Cap:", req_yq['summaryDetail']['marketCap'])
print("Beta:", req_yq['summaryDetail']["beta"])
```

Getting Stocks data from yf.Ticker().history()

```python
# Get the current date
current_date = datetime.now()
```

```python
start_date = f"{current_date.year}-{current_date.month - 1}-1"
end_date = f"{current_date.year}-{current_date.month}-{current_date.day - 1}"
```

```python
stocks_info = yf.Ticker("AAPL").history(start=start_date, end=end_date, auto_adjust=True)
stocks_info
```
