```python
import csv
import json
```

```python
def make_json(csvFilePath, jsonFilePath):

	data = {}

	with open(csvFilePath, encoding='utf-8') as csvf:
		csvReader = csv.DictReader(csvf)

		for rows in csvReader:
			key = rows['Symbol']
			data[key] = rows

	with open(jsonFilePath, 'w', encoding='utf-8') as jsonf:
		jsonf.write(json.dumps(data, indent=4))
```

```python
csvFilePath = r'./outputs_csv/s&p500_esg_data.csv'
jsonFilePath = r'./outputs_json/s&p500_esg_data.json'
make_json(csvFilePath, jsonFilePath)
```

```python
def make_json2(csvFilePath, jsonFilePath):
    data = {}

    with open(csvFilePath, encoding='utf-8') as csvf:
        csvReader = csv.DictReader(csvf)

        for rows in csvReader:
            date = rows['Date']
            for key, value in rows.items():
                if key == 'Date':
                    continue
                if key not in data:
                    data[key] = {}
                data[key][date] = value

    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf:
        jsonf.write(json.dumps(data, indent=4))
```

```python
csvFilePath = './outputs_csv/s&p500_price_data.csv'
jsonFilePath = './outputs_json/s&p500_price_data.json'
make_json2(csvFilePath, jsonFilePath)
```
